### 1.2 Physics

**Avalanche Breakdown**

![](https://lastminuteengineers.com/wp-content/uploads/basic/Avalanche-Effect.gif)

Avalanche Breakdown in a dielectric media is a phenomena that occur in both insulating and semiconducting materials. A free electron becomes sufficiently accelerated by an external applied electric field that it fres up the additional electrons as it collides with atoms or molecues of dielectric media. Then the secondary electrons are also accelerated, creating larger numbers of free electrons. The resulting exponential growth of the electrosn and ions may rapidly lead to complete dielectric breakdown of the material. This is a very surprising phenomena where very large currents can flow even in good insulators .
