#### Mutable vs Immutable

Julia variables are two types, mutable and immutable. Mutable means that the values aof your data can be changed in-place 

````julia
a_array = ["navy", "kusam"]
a_tuple = ("navy", "kusam")
````

````julia
ismutable(a_array)
````

````julia
ismutable(a_tuple)
````

Julia passes vlues by reference, this means that if a utable object is givn to a function, and this object is mutated inside the function, the final result is kept at the passsed object

````julia
function Capslock!(x)
    x[1] = "NAVY"
    x[2] = "KUSAM"
    return x
end
````

````julia
Capslock!(a_array)
````

````julia
a_array
````

note that by pasing a_array to the function Capslock! it have updated the variable a_array

BY convention functions name ending with ! will change the input variables if they are mutable, (it cannot change the immutable variables)

````julia
Capslock!(a_array)
a_array
````

````julia
Capslock!(a_tuple)
a_tuple
````

## 3. Workflow 

Here we will today learn how to solve ODEs usng the `DifferentailEquations.jl`. 
The following are the main steps to solve ODE

1. Make the set of ODES as a Julia function f
2. Define an ODE problem: f, initial state, a parametr container
3. Choose solver
4. Solve the ODE using `solve` 
5. Analyze the solution

## 4. Simple Example

Here we will work with the exmples to learn how to solve ODES of the form

 $$\frac{du}{dt} = f(u,p,t) $$
 1. time:                  `t` 
 2. Parameters container:  `p`
 3. scalar of independant varaible: `u`
 

#### simple example (first order linear ODE)

let us take the example where $t \in [0,1] $ and $f(u,p,t) = \alpha u $

$$ \frac{du}{dt} = \alpha u $$ 

We already know the analytical solution form calculus which is

$$u_{t} = u_{0} e^{\alpha t} $$

we will solve this numerically using the packages in `DifferentialEquations.jl`

````julia
using OrdinaryDiffEq
````

````julia
using Plots
````

````julia
# step 1: Make set of ODEs as Julia function
f(u,p,t) = p*u              
````

````julia
# Step 2: define ode plorblen
u0    = 0.5                 # inital condition
p     = 1.01                # parameter
tspan = (0.0, 1.0)          # time span of simulation
prob  = ODEProblem(f, u0, tspan, p)  
````

````julia
# step 3: Choose algorithem as solver
alg   = Tsit5()             # 5th order Tsitouras method 
````

````julia
# step 4: Solve ODE
sol   = solve(prob, alg=alg )
````

````julia
sol.t , sol.u
````

````julia
# step 5: Visualization
plot(sol)
````

## 5. Better Example: Solving Lorentz Equations

#### Step 1: Make the set of ODES as a Julia function f


````julia
function lorentz_func(u, p, t)
    """
    ODE function
    u : [x, y ,z]
    p : [σ, ρ, β]
    t : (0, T)
    """
    x, y, z = u 
    σ, ρ, β = p
    
    ẋ = σ*(y-x)
    ẏ = x*(ρ-z)-y
    ż = x*y-β*z
    
    return [ ẋ, ẏ, ż ]
end
````

#### Step 2: Define an ODE problem: f, initial state, a parametr container

````julia
#initial state
u₀ = [10.0, 10.0, 10.0]
#parameters
p  = [10.0, 28.0, 8.0/3.0]
# time span
tspan = (0.0, 100.0)
# define ODE problem
prob = ODEProblem(lorentz_func, u₀, tspan, p)
````

#### Step 3: Choose solver

````julia
# step 3: Choose algorithem as solver
alg   =  Vern9()             # 9th order adaptive RK method 
````

#### Step 4: Solve the ODE using `solve` 

````julia
sol   = solve(prob, alg=alg )
````

#### Step 5: Analyze the solution
